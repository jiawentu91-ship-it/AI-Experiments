```python
import random

class GuessNumberGame:
    def __init__(self):
        # 初始化游戏：生成目标数字，重置猜测次数
        self.reset_game()
    
    def reset_game(self):
        # 单独的重置方法，替代直接调用__init__
        self.target = random.randint(1, 100)
        self.times = 0
    
    def guess_num(self, x):
        """
        接收猜测的数字，返回是否猜对，同时给出提示
        :param x: 玩家猜测的数字
        :return: 猜对返回True，否则返回False
        """
        self.times += 1  # 简化次数累加写法
        if x > self.target:
            print("猜大了")
            return False
        elif x < self.target:
            print("猜小了")
            return False
        else:
            print("猜对了！")
            return True
    
    def show_status(self):
        """展示当前游戏状态（次数）"""
        remaining = 7 - self.times
        print(f"已进行{self.times}次，{'还剩' if remaining > 0 else '已用完'} {max(0, remaining)}次")

# 实例化对象（类名大写，实例名小写，避免冲突）
game = GuessNumberGame()

# 游戏主循环
for _ in range(7):
    try:  # 增加异常处理，防止输入非数字
        x = int(input("请输入你猜的数字（1-100）："))
        if not 1 <= x <= 100:
            print("请输入1-100之间的数字！")
            continue
    except ValueError:
        print("输入错误！请输入有效的数字。")
        continue
    
    # 调用猜数字方法，接收是否猜对的结果
    is_correct = game.guess_num(x)
    game.show_status()
    
    if is_correct:
        print(f"恭喜你，只用了{game.times}次就猜对了！")
        game.reset_game()  # 重置游戏
        break
    # 最后一次没猜对的提示
    if game.times >= 7:
        print(f"次数已用光！正确数字是{game.target}")
        game.reset_game()
        break
```

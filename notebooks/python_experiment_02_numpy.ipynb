```python
import numpy as np
A=np.mat("1 -2 1;0 2 -8;-4 5 9")
P,Q= np.linalg.eig(A)

U, Sigma, V = np.linalg.svd(A)

b=np.array([0,8,-9])
X=np.linalg.solve(A,b)
P,Q,U,Sigma,V,X
```

```python
from pylab import*
import numpy as np
x=np.arange(-np.pi,np.pi,0.1)
subplot(1,2,1)
plot(x,sin(x),color='blue',linewidth='4',linestyle='-',label='sinx')
xlabel('x')
ylabel('y')
legend(loc='upper left')
grid(True)
subplot(1,2,2)
plot(x,cos(x),color='red',linewidth='5',linestyle='-.',label='cosx')
xlabel('x')
ylabel('y')
legend(loc='upper left')
grid(True)
```

```python
import matplotlib as mpl
from pylab import*
mpl.rcParams['font.family'] = 'sans-serif'
mpl.rcParams['font.sans-serif'] = [u'SimHei']
data=[5,10,10,65,5,5]
labels=['娱乐','育儿','饮食','房贷','交通','其他']
subplot(1,2,1)
pie(data,explode=[0,0,0,0.3,0,0],labels)
subplot(1,2,2)
left=range(6)
bar(left,height=data,width=0.3)
x=range(6)
plot(x,data,'r','2','-')
xticks(range(6), labels)
```

```python
mpl.rcParams['font.family'] = 'sans-serif'
mpl.rcParams['font.sans-serif'] = [u'SimHei']
mpl.rcParams['axes.unicode_minus'] = False
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 
fig = plt.figure() 
ax = fig.add_subplot(1,1,1, projection='3d') 
z=np.arange(0,30,0.1)
x=sin(z)
y=cos(z)
ax.plot(x,y,z,'r')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 
delta = 0.025
x = np.arange(-10, 10, delta)
y = np.arange(-10, 10, delta)
X, Y = np.meshgrid(x, y)
Z=1/(np.sqrt((1-X)**2+Y**2)+np.sqrt((1+X)**2+Y**2))
plt.contour(X,Y,Z,20)
```
